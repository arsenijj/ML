```python
def isRefl(a):
    anti_refl, refl = True, True
    for i in range(len(a)):
        if not a[i][i]:
            refl = False
        if a[i][i]:
            anti_refl = False
        if not refl and not anti_refl:
            return False, False
    return refl, anti_refl


isRefl(a)
```

```python
def isSymm(a):
    symm, anti_symm = True, True
    for i in range(len(a)):
        for j in range(len(a)):
            if a[i][j] + a[j][i] not in [0, 2] and i != j:
                symm = False
            if a[i][j] + a[j][i] == 2 and i != j:
                anti_symm = False
            if not symm and not anti_symm:
                return False, False
    return symm, anti_symm


isSymm(a)
```

```python
def isTran(a):
    for i in range(len(a)):
        for j in range(len(a)):
            for k in range(len(a)):
                if a[i][j] + a[j][k] == 2 and not a[i][k]:
                    return False
    return True

isTran(a)
```

```python
def classifier(a):
    reflexive, anti_reflexive = isRefl(a)
    symmetric, anti_symmetric = isSymm(a)
    transitive = isTran(a)
    res = []
    if transitive:
        if reflexive:
            res.append('Kvazi')
            if symmetric:
                res.append('Equivalent')
        if reflexive and anti_symmetric:
            res.append('Partial order')
        if anti_reflexive and anti_symmetric:
            res.append('Strict order')
    return res

classifier(a)
```
